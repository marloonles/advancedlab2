```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, fftfreq

path = "/content/advlab2lab1data.csv"
data = np.genfromtxt(path, delimiter=',', names='frequency, level')

freq = data['frequency']
level = data['level']

#figure code format from https://pythonnumericalmethods.studentorg.berkeley.edu/notebooks/chapter24.04-FFT-in-Python.html

plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.title('FFT Frequencies from Tuning Fork 1')
plt.plot(freq, level)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Level (dB)')
plt.grid()
plt.xlim(0,1000)

plt.subplot(122)
plt.title('Zoomed in Frequency from Tuning Fork 1')
plt.plot(freq, level)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Level (dB)')
plt.grid()
plt.xlim(894,896)
plt.ylim(-20, -10)
plt.show()
```
